```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('daily_min_temp.csv', index_col="Date", parse_dates=["Date"])
df
```

```python
df=df.resample('M').mean()
df
```

```python
cols = list()
```

```python
for i in range(1, 0, -1):
    cols.append(df.shift(i))
cols
```

```python
for i in range(2, 0, -2):
     cols.append(df.shift(i))
cols
```

```python
for i in range(3, 0, -3):
     cols.append(df.shift(i))
cols
```

```python
for i in range(0, 1):
    cols.append(df.shift(-i))
cols
```

```python
agg = pd.concat(cols, axis=1)
agg
```

```python
agg.dropna(inplace=True)
```

```python
train, test=agg.iloc[:-10, :],agg.iloc[-10:, :]
```

```python
train
```

```python
test
```

```python
trainX, trainy = train.iloc[:, :-1], train.iloc[:, -1]
```

```python
trainX
```

```python
trainy
```

```python
#modeling 
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=1000)
model.fit(trainX, trainy)
```

```python
#predicted
```

```python
from sklearn.metrics import mean_absolute_error
predictions=[]
actual=[]
for i in range(len(test)):
    testX, testy = test.iloc[i, :-1], test.iloc[i, -1]
    pred = model.predict([testX])
    print("hasil prediksi:"+str(pred[0]))
    predictions.append(pred[0])
    actual.append(testy)
    print("expected= "+str(testy)+ " predicted= "+str(pred))
```

```python
error = mean_absolute_error(actual, predictions)
print(error)
```

```python
plt.plot(actual, label='Actual')
plt.plot(predictions, label='Predicted')
plt.legend()
plt.show()
```

```python
df_visual = pd.DataFrame({
    'date':test.reset_index()['Date'].to_list(),
    'actuals':actual
})
```

```python
df_visual
```

```python
df_visual['pred']=predictions
```

```python
df_visual.set_index('date', inplace=True)
df_visual
```

```python
plt.figure(figsize=(15, 5))
plt.plot(df_visual.actuals, "b-",label='actual')
plt.plot(df_visual.pred, "r-",label='prediction')
```

```python
result = pd.concat([df, df_visual], axis=1).drop(columns='actuals')
```

```python
result
```

```python
plt.figure(figsize=(15, 5))
plt.plot(result.Temp[-50:], "b-",label='actual')
plt.plot(result.pred, "r-",label='predicted')
plt.legend()
```
