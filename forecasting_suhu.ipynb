```python
df= pd.read_csv('daily_min_temp.csv',index_col="Date",parse_dates=["Date"])
df
```

```python
df=df.resample('M').mean()
df
```

```python
df.Temp.plot(figsize=(15,5))
```

```python
# data prossing 
```

```python
cols=list() 
```

```python
for i in range(1,0,-1):
  cols.append(df.shift(i))
cols 
```

```python
for i in range(0,1):
    print(i)
    cols.append(df.shift(i))
cols
```

```python
agg=pd.concat(cols,axis=1)
agg
```

```python
agg.dropna(inplace=True)
 
```

```python
train=agg.head(109)
train 
```

```python
test=agg.tail(10)
test
```

```python
# tarinig 
```

```python
train
```

```python
trainX=train.iloc[:,:-1]
trainX
```

```python
trainy=train.iloc[:,-1]
trainy
```

```python
from sklearn.ensemble import RandomForestRegressor 
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR 
model=LinearRegression()
model.fit(trainX,trainy)



```

```python
actual=list()
predictions=list()
for i in range(len(test)):
    testX,testy=test.iloc[i,:-1],test.iloc[i,-1]
    pred=model.predict([testX])
    print("hasil prediksi :"+str(pred[0]))
    predictions.append(pred[0])
    actual.append(testy)
    print("actual: "+str(testy)+" predicted= "+str(pred[0]))
```

```python
actual
```

```python
predictions
```

```python
from sklearn.metrics import mean_absolute_error
error=mean_absolute_error(actual,predictions)
print(error)
```

```python
plt.plot(actual, label="actual")
plt.plot(predictions, label="predictions")
plt.legend()
plt.show()
```

```python
test.reset_index()["Date"].to_list()
```

```python
df_visual = pd.DataFrame({
    'date':test.reset_index()["Date"].to_list(),
    'actual':actual,
    'predictions':predictions
})
```

```python
df_visual.set_index('date',inplace=True)
```

```python
df_visual
```

```python
plt.figure(figsize=(15,5))
plt.plot(df_visual.actual,"b-",label="actual")
plt.plot(df_visual.predictions,"r-",label="predictions")
plt.legend()
plt.show()
```

```python
result=pd.concat([df,df_visual],axis=1)
```
